```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\08-Linear-Regression-Models\\Advertising.csv")
```

```python
df.head()
```

```python
df['total'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data = df, x = df.total, y = df.sales)
```

```python
sns.regplot(data = df, x = df.total, y = df.sales)
# seaborn is capable of performing simple linear regression with regplot
```

```python
X = df.total 
Y = df.sales
```

```python
#help(np.polyfit)
```

```python
np.polyfit(X,Y,deg = 1)
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales = potential_spend * 0.04868788 + 4.24302822
```

```python
sns.scatterplot(data = df, x = df.total, y = df.sales)
plt.plot(potential_spend,predicted_sales, color = 'red')
```
